```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.tri as mtri

from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
i_ev=0
i_time_todist=0.05
eps=0.10
min_samples=15
#csvlist = !ls -lrt ev_synthetic_trg_v_*.csv | awk '{print $9}' 
csvlist = !ls -lrt ev_synthetic_trg_v_4727ev.csv | awk '{print $9}'
csvlist
```

```python
csvlist[i_ev]
```

```python
df=pd.read_csv(csvlist[i_ev],header=None,sep=" ",names=['x','y','i','j','chID'])
```

```python
df['z']=df['i']*i_time_todist
```

```python
df['z'].hist(bins=100)
```

```python
ax=df.plot.scatter('x','y',figsize=(5,5));
ax.set(xlabel='x, m',ylabel='y, m');
ax.yaxis.grid(True)
ax.xaxis.grid(True)
```

```python
df[df['i']==0].plot.scatter('x','y',figsize=(5,5))
```

```python
X=np.array(df[['x','y','z']])
```

```python
clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(X)
labels = clustering.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print(len(X))
print(len(labels))
print(sum(labels)/len(clustering.labels_))
print("Estimated number of clusters     : %d" % n_clusters_)
print("Estimated number of noise points : %d" % n_noise_)
```

```python
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[clustering.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
plt.figure(figsize=(10,10))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title(f"Estimated number of clusters: {n_clusters_}")
plt.show()
```
